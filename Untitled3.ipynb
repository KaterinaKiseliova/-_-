```python
import pandas as pd


df = pd.read_csv('kc_house_data.csv')
df.head()
```

```python
import matplotlib.pyplot as plt


plt.hist(df['price'])
plt.title("Изучите стоимости недвижимости")
plt.xlabel('стоимости недвижимости')
plt.ylabel('кол-во')
```

```python
plt.hist(df['sqft_living'])
plt.title("Изучите распределение квадратуры жилой")
plt.xlabel('жилая площадь')
plt.ylabel('кол-во домов')
```

```python
plt.hist(df['yr_built'])
plt.title("Изучите распределение года постройки")
plt.xlabel('год постройки')
plt.ylabel('кол-во домов')
```
